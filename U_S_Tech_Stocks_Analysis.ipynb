<h3> Simple Moving Averages(SMA) and Exponential Moving Averages(EMA) for portfolios </h3>

1. A Simple Moving Average calculates the average of a selected range of prices, usually closing prices, by the number of periods in that range. 
2. The SMA is a technical indicator that can aid in determining if an asset price will continue or reverse a bullish or bearish trend. 
3. The SMA can be enhanced as an exponential moving average (EMA) that more heavily weights recent price action. 

```python
SP500.head()
```

```python
SP500.rolling(window = 10)      # a rolling window of 10 days
```

- pandas _dataframe.rolling()_ function provides the featur of rolling window calculations. 
- We take a window of size k at a time and perform some desired mathematical operation on it.

```python
type(SP500.rolling(window=10))
```

```python
SP500.Close.rolling(window = 10, min_periods = 5).mean()          # a simple moving average of 10 days
                                                                  # (min_period = n)> at least n valid observations
```

```python
SP500.Close.rolling(window = 10, min_periods = 5).mean().plot(figsize = (15,8), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
SMA = pd.DataFrame()

SMA['SP500_SMA10'] = norm_stocks.SP500.rolling(window = 10, min_periods = 5).mean()
SMA['EWP_tech_SMA10'] = norm_stocks.EWP_tech.rolling(window = 10, min_periods = 5). mean()

SMA
```

```python
SMA.plot(figsize = (15,8), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
SMA['SP500_SMA200'] = norm_stocks.SP500.rolling(window = 200, min_periods = 10).mean()                     
SMA['EWP_tech_SMA200'] = norm_stocks.EWP_tech.rolling(window = 200, min_periods = 10).mean()            # simple moving average of 200 days with minimum time period of 10 days for the calculation of average
```

```python
SMA
```

<h5> Exponentially-weighted Moving Averages </h5>
- Newer data points are given more importance or a greater weight and thus have more effect on the calculation of the moving average.

#### Formula: 
![Exponential Moving Average Equation Source: Investopedia](https://cdn-images-1.medium.com/max/800/1*IF5pSc_Zp_vlKI3paux4-Q.png)

In the case of inbuilt function ewm() i.e. exponential weighted moving average -> span = no. of days for which the data is observed and (smoothing = 2) is taken by default
<br> => α = 2/(1 + span)

```python
SMA['SP500_EMA200'] = norm_stocks.SP500.ewm(span = 200, min_periods = 10).mean()                     
SMA['EWP_tech_EMA200'] = norm_stocks.EWP_tech.ewm(span = 200, min_periods = 10).mean()
```

<h5> Expanding Windows </h5>
- For each timestamp expanding().mean() gives the average stock value including all preceding timestamps

```python
SMA['SP500_EXP'] = norm_stocks.SP500.expanding().mean()                     
SMA['tech_EXP'] = norm_stocks.EWP_tech.expanding().mean()
```

```python
SMA.iloc[:, -2:].plot(figsize = (15,12), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.plot(SMA['SP500_SMA10'], label = 'SMA_10', color = 'green', alpha = 0.6)
plt.plot(SMA['SP500_SMA200'], label = 'SMA_200', color = 'Orange', alpha = 0.8)
plt.plot(SMA['SP500_EMA200'], label = 'EMA_200', color = 'red', alpha = 0.6)
plt.plot(SMA['SP500_EXP'], label = 'EMA_200', color = 'blue', alpha = 0.6)
plt.legend(fontsize = 15, loc = 'upper left')
plt.title('Momentum Trading for S&P 500', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.plot(SMA['EWP_tech_SMA10'], label = 'SMA_10', color = 'green', alpha = 0.6)
plt.plot(SMA['EWP_tech_SMA200'], label = 'SMA_200', color = 'Orange', alpha = 0.8)
plt.plot(SMA['EWP_tech_EMA200'], label = 'EMA_200', color = 'red', alpha = 0.6)
plt.plot(SMA['tech_EXP'], label = 'EMA_200', color = 'blue', alpha = 0.6)
plt.legend(fontsize = 15, loc = 'upper left')
plt.title('Momentum Trading for Tech Stocks', fontsize = 20)
plt.show()
```

<h3> Performance Reporting with rolling() </h3>

- In finance it is common to report annual risk and return on the basis of monthly risk and return for last 3 years (36 month)

```python
stocks_1.EWP_tech.resample("M", kind = "period").last()
```

```python
stocks_1.SP500.resample("M", kind = "period").last()
```

```python
month_ret_tech = pd.DataFrame()

month_ret_tech['Close'] = stocks_1.EWP_tech.resample("M", kind = "period").last().pct_change().dropna()
month_ret_tech['Return'] = month_ret_tech['Close'].rolling(18).mean()*12        # getting annual return on the basis of monthly return for last 1.5 years
month_ret_tech['Risk'] = month_ret_tech['Close'].rolling(18).std()*np.sqrt(12)  # getting annual risk on the basis of monthly return for last 1.5 years

month_ret_tech.dropna(inplace = True)
month_ret_tech
```

```python
month_ret_SP500 = pd.DataFrame()

month_ret_SP500['Close'] = stocks_1.SP500.resample("M", kind = "period").last().pct_change().dropna()
month_ret_SP500['Return'] = month_ret_SP500['Close'].rolling(18).mean()*12
month_ret_SP500['Risk'] = month_ret_SP500['Close'].rolling(18).std()*np.sqrt(12)

month_ret_SP500.dropna(inplace = True)
month_ret_SP500
```
